# Split & Merge

## Conceptos:

|   |   |    |
|---|---|---|
|Codigo | Concepto | Descripció|
| Split() | Separación   | Separación de canales, este proceso, devuelve 3 valores (Red, Blue, Green) |
| Merge() | Union   | Este proceso unifica canales establecidos (Como lo pueden ser RGB)|


## Importing Libraries <a class="anchor" id="libraries" ></a>

```python
import cv2
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import glob
from google.colab.patches import cv2_imshow
from IPython.utils.py3compat import iteritems
```

Imagen Original a procesar:

image1.jpg

```python
img = cv2.imread('/content/data/image1.jpg',cv2.IMREAD_UNCHANGED)

b,g,r = cv2.split(img)
merged = cv2.merge((b,g,r))

titulos = ['Blue-Split','Green-Split','Red-Split','Merged Image']
imagenes = [b,g,r,merged]
figure(figsize=(6, 4), dpi=200)

for i in range (4):
  plt.subplot (2,2,i+1),plt.imshow(cv2.cvtColor(imagenes[i],cv2.COLOR_BGR2RGB))
  plt.title(titulos[i])
  plt.xticks([]),plt.yticks([])
plt.show()
```

## Conclusiones:

* *Con -SPLIT- herramienta podemos realizar una separación de los canales de una imagen, en este ejercicio podemos observar como la instrucción nos devuelve los 3 canales y la podemos mostrar posteriormente unificada con la instrucción -MERGE- *
