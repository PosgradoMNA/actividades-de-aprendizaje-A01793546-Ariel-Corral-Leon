# Algoritmo de Canny

## Conceptos:


Al realizar el procesamiento de una imagen a travez de la función Canny se realizan 5 pasos:

* *Noise reduction*
* *Gradient calculation*
* *Non-maximum suppression*
* *Double threshold*
* *Edge Tracking by hysterisis*

Posterior a esto, nuestra imagen ya denota bordes y estos se pueden acentuar o suavizar mediante el estar realizando ajustes en los Trhresholds que son parte de las funciones:

## Importing Libraries <a class="anchor" id="libraries" ></a>

```python
import cv2
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import glob
from google.colab.patches import cv2_imshow
from IPython.utils.py3compat import iteritems
```

Imagen Original a procesar:

```python
img1 = cv2.imread('/content/data/image1.jpg',cv2.IMREAD_UNCHANGED)
img2 = cv2.imread('/content/data/image2.jpg',cv2.IMREAD_UNCHANGED)
img3 = cv2.imread('/content/data/image3.jpg',cv2.IMREAD_UNCHANGED)
img4 = cv2.imread('/content/data/image4.jpg',cv2.IMREAD_UNCHANGED)

canny1 = cv2.Canny(img1,300 ,300)
canny2 = cv2.Canny(img2,100 ,500)
canny3 = cv2.Canny(img3,100 ,100)
canny4 = cv2.Canny(img4,100 ,500)

titulos = ['Original Image #1','Canny #1','Original Image #2','Canny #2','Original Image #3','Canny #3','Original Image #4','Canny #4']
imagenes = [img1,canny1,img2,canny2,img3,canny3,img4,canny4,]
figure(figsize=(8, 18), dpi=200)

for i in range (8):
  plt.subplot (4,2,i+1),plt.imshow(cv2.cvtColor(imagenes[i],cv2.COLOR_BGR2RGB))
  plt.title(titulos[i])
  plt.xticks([]),plt.yticks([])
plt.show()
```

## Conclusiones:

* *Con Canny, podemos obtener bordes de una imagen o un vidéo ya que realiza diversos procesos para poder identificar las partes con contrastes de tal manera que logra identificarse los bordes*
