## Importing Libraries <a class="anchor" id="libraries" ></a>

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import cv2
import glob
import pandas as pd
#from google.colab.patches import cv2_imshow
from IPython.utils.py3compat import iteritems
from skimage import data
from skimage.filters import threshold_otsu
from skimage.filters import threshold_multiotsu
from skimage.filters import try_all_threshold
```

Imagen Original a procesar:

```python
img1 = cv2.imread('/content/data/image1.png', cv.IMREAD_GRAYSCALE)

th = int(input("Introduce el valor de Threshold en valor entero: "))
max_val = 255

ret, imagen1 = cv2.threshold(img1, th, max_val, cv2.THRESH_BINARY + cv2.THRESH_OTSU )
ret, imagen2 = cv2.threshold(img1, th, max_val, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU )
ret, imagen3 = cv2.threshold(img1, th, max_val, cv2.THRESH_TOZERO + cv2.THRESH_OTSU )
ret, imagen4 = cv2.threshold(img1, th, max_val, cv2.THRESH_TOZERO_INV + cv2.THRESH_OTSU )
ret, imagen5 = cv2.threshold(img1, th, max_val, cv2.THRESH_TRUNC + cv2.THRESH_OTSU )

titulos = ['Original','Bin + OTSU','Bin Inv + OTSU','Tozero + OTSU','2Zero Inv + OTSU','Trunc + OTSU', 'OTSU']
imagenes = [img1,imagen1, imagen2,imagen3,imagen4,imagen5]
figure(figsize=(8, 18), dpi=300)

for i in range (6):
  plt.subplot (2,3,i+1)
  plt.imshow(imagenes[i], cmap=plt.cm.gray)
  plt.title(titulos[i])
  plt.xticks([]),plt.yticks([])
plt.show()
```
